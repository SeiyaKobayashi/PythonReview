```python
class Node:
    def __init__(self, val=None):
        self.val  = val
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def printLinkedList(self):
        curr = self.head
        while curr != None:
            print(curr.val, end=' ')
            curr = curr.next
    
    def insertHead(self, data):
        newNode = Node(data)
        newNode.next = self.head
        self.head = newNode
        
    def insertTail(self, data):
        newNode = Node(data)
        
        if self.head is None:
            self.head = newNode
            return
        
        tail = self.head
        while(tail.next):
            tail = tail.next
        tail.next = newNode
    
    def insertMiddle(self, nodeBefore, data):
        if nodeBefore is None:
            print("Invalid node.")
            return

        newNode = Node(data)
        newNode.next = nodeBefore.next
        nodeBefore.next = newNode
        
    def removeNode(self, key):
        head = self.head
        
        if (head == None):
            print('Linked list is empty. Nothing can be removed.')
            return

        if (head is not None):
            if (head.val == key):
                self.head = head.next
                head = None
                return
            else:
                while (head is not None):
                    if head.val == key:
                        break
                    prev = head
                    head = head.next
                else:
                    print('Invalid key.')
                    return
        
        prev.next = head.next
        head = None

linkedList_1 = LinkedList()

linkedList_1.head = Node(10)
e2 = Node(20)
e3 = Node(30)
e4 = Node(40)

linkedList_1.head.next = e2
e2.next = e3
e3.next = e4

linkedList_1.insertHead(0)
linkedList_1.insertTail(50)
linkedList_1.insertMiddle(linkedList_1.head.next.next, 25)
linkedList_1.removeNode(20)

linkedList_1.printLinkedList()
```
