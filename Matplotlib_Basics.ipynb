# Matplotlib Basics

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Simple Plot

```python
# Data to plot
x = np.linspace(0, 101, 100)
y = np.random.randint(0, 101, 100)

# Plot a graph
plt.plot(x, y, label="sample")

# Add legend
plt.legend()

# Show the graph
plt.show()
```

```python
# Data to plot
x = np.linspace(0, 101, 100)
y = np.random.randint(0, 101, 100)

# Initialize figure
fig = plt.figure()

# Add axes
ax = fig.add_subplot(111)
ax.plot(x, y, label="sample_2")

# Add legend
plt.legend()

# Show the graph
plt.show()
```

## 2. Figure & Axes

```python
# Data to plot
x = np.array([1, 2, 3, 4, 5])
y = np.absolute(np.random.randn(5))

# Initialize figure
fig = plt.figure(figsize=(10, 4))
ax_1 = fig.add_subplot(121)
ax_2 = fig.add_subplot(122)

# Plot bar graphs
ax_1.bar(x, y, label="sample_3")
ax_2.barh(x, y, label="sample_4")

# Add legends
ax_1.legend()
ax_2.legend()

# Show the graph
plt.show()
```

```python
# Data to plot
x = np.array([1, 2, 3, 4, 5])
y = np.absolute(np.random.randn(5))

# Initialize figure & axes
fig, (ax_1, ax_2) = plt.subplots(1, 2, figsize=(10, 4))

# Plot bar graphs
ax_1.bar(x, y, label="sample_5")
ax_2.barh(x, y, label="sample_6")

# Add legends
ax_1.legend()
ax_2.legend()

# Show the graph
plt.show()
```

## 3. Plotting Graphs

```python
# Data to plot
x = np.arange(0, 10, 0.5)
a = np.random.rand(30)
b = np.random.rand(30)
colors = np.random.rand(30)
area = (30 * np.random.rand(30))**2

# Initialize figure & axes
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plot graphs
axes[0].plot(x, x, 'r--', x, x**2, 'bs', x, x**3, 'g^')
axes[1].scatter(a, b, s=area, c=colors, alpha=0.5)

# Show the graphs
plt.show()
```

```python
# Data to plot
x = np.arange(1, 11)
y = np.random.randint(0, 101, 10)

plt.title('Sample Title')
plt.xlabel('xlabel')
plt.ylabel('ylabel')

# Plot a graph
plt.bar(x, y, label="sample_7")

# Add legend
plt.legend()

# Show the graph
plt.show()
```
